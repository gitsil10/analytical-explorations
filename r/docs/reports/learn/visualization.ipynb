```r
#data generation
url <- "https://query1.finance.yahoo.com/v7/finance/download/GOOGL?period1=1679443200&period2=1711065600&interval=1d&events=history&includeAdjustedClose=true"
df <- read.csv(url)
df$Date <- as.Date(df$Date)
googl_df <- df[, c("Date", "Close", "High", "Low")]
head(googl_df)
```

```r
#data summary
summary(googl_df)
```

<h2>Histogram</h2>
<ul>
<li>creates a frequency distribution graph</li>
<li>estimate of the probability distribution of a continuous variable</li>
<li>created with hist()</li>
</ul>

```r
#histogram
#creates a histogram
#hist(x)
#x is the data you want to plot
#represent the frequency of the data
#shows the distribution of the data

hist(
  googl_df$Close,
  main = "Histogram of Google Stock Close Price",
  xlab = "Close Price",
  ylab = "Frequency",
  col = "lightblue",
  border = "black"
)
```

```r
#barplot
#creates a bar graph
#barplot(height, width = 1, space = NULL, names.arg = NULL)
#represents the data in the form of bars
#height: a vector or matrix of values describing the bars which make up the plot
#width: a scalar or vector of values describing the width of the bars
#space: a scalar or vector of values describing the space between the bars
#each bar represents a value of the vector
#requires a vector of values to plot
#discusses the data in terms of the number of occurrences of each value
#useful for comparing the frequency of different values

barplot(
  googl_df$Close,
  main = "Barplot of Google Stock Close Price",
  xlab = "Date",
  ylab = "Close Price",
  col = "lightblue",
  border = "black"
)
```

```r
#scatter.smooth
#creates a scatter plot with a smooth line
#scatter.smooth(x, y, main, xlab, ylab, col)
#x = x-axis data
#y = y-axis data
#shows the relationship between two numerical variables
#each value is a point on the graph
#the smooth line is a line that shows the general trend of the data
#identifies patterns in the data
#helps to predict future values
#depicts correlation between two variables

scatter.smooth(
  googl_df$Low,
  googl_df$Close,
  main = "Scatterplot of Google Stock Close Price",
  xlab = "Low",
  ylab = "Close Price",
  col = "blue"
)
```

```r
#plot
#creates a line graph
#x-axis is the independent variable
#y-axis is the dependent variable
#type = "l" specifies that the graph should be a line graph
#discusses the relationship between two variables
#shows the trend of the data

plot(
  googl_df$Date,
  googl_df$Close,
  main = "Google Stock Close Price Over Time",
  xlab = "Date",
  ylab = "Close Price",
  type = "l",
  col = "blue"
)
```

```r
#curve
#creates a curve graph
#curve(expr, from, to, col, lwd, main, xlab, ylab)
#expr: an expression that defines the curve
#from: the starting point of the curve
#to: the ending point of the curve
#col: the color of the curve
#lwd: the width of the curve
#represent the data in the form of a curve
#shows the relationship between two variables

curve(
  dnorm(x, mean = mean(googl_df$Close), sd = sd(googl_df$Close)),
  from = min(googl_df$Close),
  to = max(googl_df$Close),
  col = "blue",
  lwd = 2,
  main = "Density Plot of Google Stock Close Price",
  xlab = "Close Price",
  ylab = "Density"
)
```

```r
#pie
#creates a pie chart
#pie(x, labels, main, col)
#x: a vector of non-negative numerical values
#represent the data in the form of a pie chart
#shows the proportion of each value in the data
#useful for comparing the distribution of different values

vals <- c(mean(googl_df$Close), mean(googl_df$High), mean(googl_df$Low))
pie(
  vals,
  labels = c("Close", "High", "Low"),
  main = "Pie Chart of Google Stock Close Price",
  col = rainbow(length(vals))
)
```

```r
#boxplot
#creates a box plot
#boxplot(x, main, xlab, ylab, col, border)
#x: a vector or matrix of values describing the data to be plotted
#represent the data in the form of a box plot
#shows the distribution of the data
#useful for comparing the spread of different values
#identifies outliers in the data
#shows the median, quartiles, and range of the data

for (i in 2:ncol(googl_df)) {
  boxplot(
    googl_df[, i],
    main = paste("Boxplot of", colnames(googl_df)[i]),
    xlab = colnames(googl_df)[i],
    ylab = "Close Price",
    col = "lightblue",
    border = "black"
  )
}
```
